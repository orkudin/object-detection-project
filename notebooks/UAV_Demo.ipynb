# 📈 UAV-CV-Navigator: Анализ телеметрии (Google Colab)
Данный ноутбук предназначен для валидации производительности пайплайна обнаружения и трекинга (построенного для Ubuntu 20.04/22.04).

**Инструкция:**
1. Запустите пайплайн на вашем ПК: `python src/pipeline.py`
2. Возьмите сгенерированный файл `data/output/telemetry.jsonl`
3. Загрузите этот файл прямо сюда, в боковую панель файлов Google Colab
4. Запустите ячейку ниже для получения графиков метрик по ТЗ.

```python
import pandas as pd
import matplotlib.pyplot as plt
import json
import os

if not os.path.exists('telemetry.jsonl'):
    print("ОШИБКА: Пожалуйста загрузите файл telemetry.jsonl в файлы Colab!")
else:
    data = []
    with open('telemetry.jsonl', 'r') as f:
        for line in f:
            data.append(json.loads(line))
            
    df = pd.DataFrame(data)
    df['obj_count'] = df['tracks'].apply(len)
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    
    # График 1: FPS (Производительность)
    ax1.plot(df['frame_id'], df['fps'], color='green', linewidth=2)
    ax1.set_title('Производительность конвейера (FPS)')
    ax1.set_xlabel('Кадр (Frame ID)')
    ax1.set_ylabel('Кадров в секунду')
    ax1.grid(True, linestyle='--', alpha=0.7)
    
    # График 2: Количество объектов (Трафик)
    ax2.plot(df['frame_id'], df['obj_count'], color='purple', linewidth=2)
    ax2.set_title('Плотность фиксации объектов (Горячие точки)')
    ax2.set_xlabel('Кадр (Frame ID)')
    ax2.set_ylabel('Количество треков на кадре')
    ax2.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()
```
